```python
def model(x, t):
    return x[0] + x[1]*t + x[2]*t**2
```

```python
def func(x, data, dt):
    t = np.arange(0, len(data)*dt, dt)
    y = model(x, t)
    return 1./len(data)*np.sum((y-data)**2)
```

```python
def grad_func(x, data, dt):
    t = np.arange(0, len(data)*dt, dt)
    y = model(x, t)
    dif = y - data
    grad = np.ones(len(x))*1./len(data)*2
    grad[0] *= np.sum(dif)
    grad[1] *= np.sum(np.multiply(dif, t))
    grad[2] *= np.sum(np.multiply(dif, t**2))
    return grad
```

```python
#Define the model equations, parameters, and variables. 
def IFFL1(y, t, v):
    '''INPUTS
        y: vector of the state variables
        y(1) - ZFa RNA
        y(2) - ZFa Protein
        y(3) - ZFi-DsRed RNA
        y(4) - ZFi-DsRed Protein
        y(5) - Reporter RNA
        y(6) - Reporter Protein
        
        v = b, m, w, ZFa_dose, ZFi_dose
        '''
    #DEFINITION OF PARAMETERS
    #A constitutive CMV or CAG promoter controls ZF transcription.
    #Value based on parameter fitting to experimental data (pg. 57 Model
    #Development)
    #in arbitrary transcription/translation units
    k_txn_con = 1;
    
    #A ZF-TF controlled promoter regulated reporter transcription
    #Value based on parameter fitting to experimental data. (pg. 57 Model
    #Development)
    #in arbitrary transcription/translation units
    k_txn_rep = 9;
    
    #in arbitrary transcription/translation units
    k_trans = 1;
    
    kdeg_RNA = 2.7; #hr^-1
    kdeg_ZF = 0.35; #hr^-1
    kdeg_Reporter = 0.029; #hr^-1 0.029
#     ZFa_dose = 200
    l = 0.5
    u = 2
#     d = Reporter_plasmid_dose/200; #200 ng. (This value was chosen becaue it is the dose used to fit parameters)
    b, m, w, ZFa_dose, ZFi_dose, k_txn_end = v
    #DIFFERENTIAL EQUATIONS
    #Define the "fractional activation" of the promoter by the TF
    y1, y2, y3, y4, y5, y6 = y
    
    num1 = b + m * w * y2
    den1 = 1 + w * y2
    f1 = num1 / den1
    
    num2 = b + max((min((((4 * w * y4) / (1e-10 + w * y2) - 4 * l) * (1 - m) / (4 * u - 4 * l) + m), m)), 1) * w * y2   
    den2 = 1 + w * y2 + 4 * w * y4
    f2 = num2 / den2
    
    dydt = [k_txn_end * ZFa_dose - kdeg_RNA * y1,
            k_trans * y1 - kdeg_ZF * y2,
            k_txn_con * ZFi_dose * f1 - kdeg_RNA * y3,
            k_trans * y3 - kdeg_ZF * y4,
            k_txn_rep * f2 - kdeg_RNA * y5,
            k_trans * y5 - kdeg_Reporter * y6]
    
    return dydt

def solve_IFFL(dose_ZFa_list, dose_ZFi_list, p_list):
    #Set initial conditions and time
    y0 = [0,0,0,0,0,0]

    dt = 0.25
    t = np.arange(0, 48.25, dt)

    ZF6 = [0.02, 58, 0.043]
    b = ZF6[0]
    m = ZF6[1]
    w = ZF6[2]

    IFFL_tc_off = {}
    IFFL_tc_on = {}
    
    for i in dose_ZFi_list:        
        for j in dose_ZFa_list:
            v = b, m, w, j, i, p_list[0]
            solution_tc = odeint(IFFL1, y0, t, args=(v,))            
            IFFL_tc_off['dose ZFa='+str(j)+', dose ZFi='+str(i)] = solution_tc[:, -1]

            v = b, m, w, j, i, p_list[1]
            solution_tc2 = odeint(IFFL1, y0, t, args=(v,))            
            IFFL_tc_on['dose ZFa='+str(j)+', dose ZFi='+str(i)] = solution_tc2[:, -1]
            
    return t, IFFL_tc_off, IFFL_tc_on

dose_ZFa_list = np.arange(10, 210, 10)
dose_ZFi_list = np.arange(10, 210, 10)
p_list = [3.1, 4.1]

t, IFFL_tc_off, IFFL_tc_on = solve_IFFL(dose_ZFa_list, dose_ZFi_list, p_list)


doses_opt = []
FR_rep = [] 
for i in dose_ZFi_list:
    for j in dose_ZFa_list:
        reporter = IFFL_tc_on['dose ZFa='+str(j)+', dose ZFi='+str(i)]
        if max(reporter) > reporter[-1]:
            doses_opt.append([j, i])
            FR_rep.append([max(reporter)/reporter[-1]])

d = {'Opt Doses': doses_opt, 'FR_Rep': FR_rep}
df = pd.DataFrame(data=d)
df.sort_values('FR_Rep', ascending=False, inplace=True)


plt.plot(t, IFFL_tc_on['dose ZFa=60, dose ZFi=10'], label='dose ZFa=60, dose ZFi=10')
plt.xlabel('Time (hours)')
plt.ylabel('Reporter Expression (au)')
plt.legend()
plt.show()
```

```python
class Peak:
    def __init__(self, startidx):
        self.born = self.left = self.right = startidx
        self.died = None

    def get_persistence(self, seq):
        return float("inf") if self.died is None else seq[self.born] - seq[self.died]

def get_persistent_homology(seq):
    peaks = []
    # Maps indices to peaks
    idxtopeak = [None for s in seq]
    # Sequence indices sorted by values
    indices = range(len(seq))
    indices = sorted(indices, key = lambda i: seq[i], reverse=True)

    # Process each sample in descending order
    for idx in indices:
        lftdone = (idx > 0 and idxtopeak[idx-1] is not None)
        rgtdone = (idx < len(seq)-1 and idxtopeak[idx+1] is not None)
        il = idxtopeak[idx-1] if lftdone else None
        ir = idxtopeak[idx+1] if rgtdone else None

        # New peak born
        if not lftdone and not rgtdone:
            peaks.append(Peak(idx))
            idxtopeak[idx] = len(peaks)-1

        # Directly merge to next peak left
        if lftdone and not rgtdone:
            peaks[il].right += 1
            idxtopeak[idx] = il

        # Directly merge to next peak right
        if not lftdone and rgtdone:
            peaks[ir].left -= 1
            idxtopeak[idx] = ir

        # Merge left and right peaks
        if lftdone and rgtdone:
            # Left was born earlier: merge right to left
            if seq[peaks[il].born] > seq[peaks[ir].born]:
                peaks[ir].died = idx
                peaks[il].right = peaks[ir].right
                idxtopeak[peaks[il].right] = idxtopeak[idx] = il
            else:
                peaks[il].died = idx
                peaks[ir].left = peaks[il].left
                idxtopeak[peaks[ir].left] = idxtopeak[idx] = ir
#     return peaks
    # This is optional convenience
    return sorted(peaks, key=lambda p: p.get_persistence(seq), reverse=True)
```
